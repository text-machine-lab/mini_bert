```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import glob
```

```python
all_files = glob.glob('*.csv')
```

```python
all_files
```

```python
eval_files = [i for i in all_files if '_eval_' in i]
test_files = [i for i in all_files if '_test_' in i]
```

```python
df_list = []
for file in eval_files:
    df_i = pd.read_csv(file)
    df_i = df_i.loc[df_i.loc[:, "Embedding parameters"] > 0, :]
    df_list.append(df_i)

#
df_eval = pd.concat(df_list)
print(df_eval.shape)

#
df_list = []
for file in test_files:
    df_i = pd.read_csv(file)
    df_i = df_i.loc[df_i.loc[:, "Embedding parameters"] > 0, :]
    df_list.append(df_i)

#
df_test = pd.concat(df_list)
print(df_test.shape)
```

```python
df_eval.groupby(['run_name']).count()
```

```python
df_test.groupby(['run_name']).count()
```

```python
df_eval.to_csv("EVAL_RESULTS_ModelConfig.csv")
df_test.to_csv("TEST_RESULTS_ModelConfig.csv")
```
