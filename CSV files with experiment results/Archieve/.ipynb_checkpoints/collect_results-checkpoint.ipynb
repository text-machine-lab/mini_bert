```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import glob
```

```python
df_all = []

df_e1 = pd.read_csv("EVAL_RESULTS_ModelConfig.csv")
df_t1 = pd.read_csv("TEST_RESULTS_ModelConfig.csv")

#
df_e1["Experiment type"] = "Model config."
df_t1["Experiment type"] = "Model config."
```

```python
all_files = glob.glob('./ISO-PR/*.csv')
all_files += glob.glob('./ISO-FLOP/*.csv')
```

```python
all_files
```

```python
eval_files = [i for i in all_files if '_eval_' in i]
test_files = [i for i in all_files if '_test_' in i]
```

```python
df_list = []
for file in eval_files:
    df_i = pd.read_csv(file)
    if "ISO-PR" in file:
        df_i["Experiment type"] = "Iso-PR"
    else:
        df_i["Experiment type"] = "Iso-FLOPs"
    df_i = df_i.loc[df_i.loc[:, "Embedding parameters"] > 0, :]
    df_list.append(df_i)

#
df_eval = pd.concat(df_list + [df_e1])
print(df_eval.shape)

#
df_list = []
for file in test_files:
    df_i = pd.read_csv(file)
    if "ISO-PR" in file:
        df_i["Experiment type"] = "Iso-PR"
    else:
        df_i["Experiment type"] = "Iso-FLOPs"
    df_i = df_i.loc[df_i.loc[:, "Embedding parameters"] > 0, :]
    df_list.append(df_i)

#
df_test = pd.concat(df_list + [df_t1])
print(df_test.shape)
```

```python
df_eval.to_csv("EVAL_RESULTS.csv")
df_test.to_csv("TEST_RESULTS.csv")
```
